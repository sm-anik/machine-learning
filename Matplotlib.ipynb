# import matplot

```python
import matplotlib.pyplot as plt
import numpy as np 
 
```

```python
## practice some visual plot with two series

x=np.array([1,2,3])
y=np.array([2,4,9])
plt.plot(x,y)
```

```python
## just point out with specific character
plt.plot(x,y,'x')

## some of marker
Marker	Description
'o'	Circle	
'*'	Star	
'.'	Point	
','	Pixel	
'x'	X	
'X'	X (filled)	
'+'	Plus	
'P'	Plus (filled)	
's'	Square	
'D'	Diamond	
'd'	Diamond (thin)	
'p'	Pentagon	
'H'	Hexagon	
'h'	Hexagon	
'v'	Triangle Down	
'^'	Triangle Up	
'<'	Triangle Left	
'>'	Triangle Right	
'1'	Tri Down	
'2'	Tri Up	
'3'	Tri Left	
'4'	Tri Right	
'|'	Vline	
'_'	Hline
```

# modifying the plot

```python
## size size declare
plt.plot(x,y,marker='o', ms= 20)
```

```python
## modifying the circle colour of the previous code
plt.plot(x,y,marker='o',ms=20,mfc='g') # 'g' for colour green colour

```

```python
## modifying the border colour

plt.plot(x,y,marker='o',ms=20,mfc='g',mec='r') #'r' for red
```

```python
##Create Labels for a Plot
plt.plot(x,y)
plt.xlabel("Average pulse")
plt.ylabel("Calorie Burnage")
plt.show()
```

```python
## Creat a title for previous graph
plt.plot(x,y)
plt.xlabel("Average pulse")
plt.ylabel("Calorie Burnage")
plt.title("Sports watch data")
plt.show()
```

```python
## add grid in the background

plt.plot(x,y)
plt.xlabel("Average pulse")
plt.ylabel("Calorie Burnage")
plt.grid()
plt.show()
```

```python
## modifying the grid

plt.plot(x,y)
plt.xlabel("Average pulse")
plt.ylabel("Calorie Burnage")
plt.grid(color = 'green', linestyle = '--', linewidth = 0.5)
plt.show()
```

```python
## Create multiple plot
#plot 1:
x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])

plt.subplot(1, 2, 1)
plt.plot(x,y)

#plot 2:
x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 30, 40])

plt.subplot(1, 2, 2)
plt.plot(x,y)

plt.show()
```

```python
## add title
#plot 1:
x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])

plt.subplot(1, 2, 1)
plt.plot(x,y)
plt.title("income")

#plot 2:
x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 30, 40])

plt.subplot(1, 2, 2)
plt.plot(x,y)
plt.title("expenditure")
plt.show()
```

# pie chart

```python
pie=np.array([10,40,15,35])
plt.pie(pie)
plt.show()
```

```python
## adding labels
mylabels=["orange","apple","banana","guava"]
plt.pie(pie,labels=mylabels)
plt.show()
```

```python
## we can add starting angle
mylabels=["orange","apple","banana","guava"]
plt.pie(pie,labels=mylabels,startangle=90) #starting angle is 90
plt.show()
```

```python
## wedges to stand out
mylabels=["orange","apple","banana","guava"]
myexplode=[.2,0,0,0]
plt.pie(pie,labels=mylabels,explode=myexplode)
plt.show()
```

```python
## We can add shadow

mylabels=["orange","apple","banana","guava"]
myexplode=[.2,0,0,0]
plt.pie(pie,labels=mylabels,explode=myexplode,shadow=True)
plt.show()
```

```python
## Adding Lengend ########
mylabels=["orange","apple","banana","guava"]
myexplode=[.2,0,0,0]
plt.pie(pie,labels=mylabels,explode=myexplode)
plt.legend()
plt.show()
```

```python
## Adding title to the legend ###
mylabels=["orange","apple","banana","guava"]
myexplode=[.2,0,0,0]
plt.pie(pie,labels=mylabels,explode=myexplode)
plt.legend(title="Fruite List")
plt.show()
```
