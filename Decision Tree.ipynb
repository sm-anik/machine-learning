```python
import pandas as pd
df= pd.read_csv('F:\poralekha\programming\python\dataset\loan.csv')
print(df.info())
```

```python
print(df)
```

```python
## importing important library

from sklearn import tree
from sklearn.tree import DecisionTreeClassifier
import matplotlib.pyplot as plt
```

```python
## Creating dummy variable
df_encoded = pd.get_dummies(df, columns=[ 'marital_status','gender','education_level','occupation'], dtype = int)
print(df_encoded)
```

```python
## define feature as 'x' and instense as 'y'
x = df_encoded.drop('loan_status', axis = 1)
y = df_encoded['loan_status']
print(y)
```

```python
## split the dataset into train and test dataset

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=.1, random_state=42)
dtree= DecisionTreeClassifier()
dtree= dtree.fit(X_train, y_train)
```

```python
from sklearn.tree import DecisionTreeClassifier, plot_tree
def plot_decision_tree(model, feature_names, class_names):
    tree = plot_tree(model, feature_names = feature_names, class_names = class_names,
                     rounded = True, proportion = True, precision = 2, filled = True, fontsize=10)
    
    return tree

plt.figure(figsize=(20,10))
plot_decision_tree(dtree, x.columns, dtree.classes_)
plt.show()
```
