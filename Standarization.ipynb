```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('F:\poralekha\programming\python\dataset\Social_Network_Ads.csv')
```

```python
df.sample(5)
```

# Train test split

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(df.drop('Purchased', axis = 1), df['Purchased'], test_size=.3 , random_state=0)
```

```python
x_train.shape, x_test.shape
```

# Standard scaler

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
## Now i am going to fit the train set , it will learn the parameter from it
scaler.fit(x_train)

## Now I am going to trainsform the train and test datset for x

x_train_scaled = scaler.transform(x_train)
x_test_scaled = scaler.transform(x_test)
```

```python
scaler.mean_
```

```python
x_train_scaled = pd.DataFrame(x_train_scaled , columns = x_train.columns)
x_test_scaled = pd.DataFrame(x_test_scaled , columns = x_test.columns)
```

# Effect of scaling

```python
fig, (ax1,ax2) = plt.subplots(ncols = 2 , figsize = (12,5))

ax1.scatter(x_train['Age'],x_train['EstimatedSalary'])
ax1.set_title("Before scaling")

ax2.scatter(x_train_scaled['Age'],x_train_scaled['EstimatedSalary'],color = 'red')
ax2.set_title("After scaling")

plt.show()
```
